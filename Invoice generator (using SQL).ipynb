***Program for menu edititng and bill printing for a canteen***

# **CODE**

```python
#CREATING TABLE

import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()
c.execute('''CREATE TABLE IF NOT EXISTS MENU(CODE INT PRIMARY KEY,
ITEM VARCHAR(20), PRICE INT);''')
many_items = [(1012,"Pizza", 100),
        (1013,"Noodles", 130),
        (1014,"Burger", 150),
        (1015,"Pasta", 120),
        (1016,"Coffee", 100),
        (1017,"Salad", 90),
        (1018,"Popcorn", 120),
        (1019,"Tomato Soup", 130),
        (1020,"Icecream", 100),
        (1021,"Garlic Bread", 150)]
c.executemany("INSERT INTO MENU VALUES(?,?,?)", many_items)
conn.commit()
#c.close()
```

```python
#DISPLAY TABLE

def display():
    print("CODE".ljust(10, " "), "Item".ljust(15, " "), "Price".ljust(35, " "))
    c.execute("SELECT * FROM MENU")
    items = c.fetchall()
    for i in items:
        print(str(i[0]).ljust(10, " "), str(i[1]).ljust(15, " "), str(i[2]).ljust(35, " "))

    conn.commit()
```

```python
#ADDING ITEM TO MENU

"""import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()"""
def additem():
  try:
    sno = int(input("Code of item:"))
    c.execute("SELECT COUNT(CODE) FROM MENU WHERE CODE="+str(sno))
    a = c.fetchall()
    if a[0][0] != 0:
      print("Item existing in menu already.")
    else:
      addeditem = input("Item to add:")
      addeditem_price = int(input("Price of item:"))
      added_item = [(sno, addeditem, addeditem_price)]
      c.executemany("INSERT INTO MENU VALUES(?,?,?)", added_item)
      print("Item added successfully!")
  except:
    print("Please enter the right code")
    conn.commit()
```

```python
#DELETING ITEM FROM MENU

import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()
def deleteitem():
  try:
    delete_item = int(input("Code of item to delete:"))
    c.execute("SELECT COUNT(CODE) FROM MENU WHERE CODE="+str(delete_item))
    a = c.fetchall()
    if a[0][0] == 0:
      print("Item not in menu.")
    else:
      c.execute("DELETE FROM Menu WHERE CODE = "+str(delete_item))
      print("Item deleted successfully!")
  except:
    print("Please enter the right code")
    conn.commit()
```

```python
#MODIFYING ITEM FROM TABLE

import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()
def modifyitem():
  try:
    modify_item = int(input("Code of item to modify:"))
    c.execute("SELECT COUNT(CODE) FROM MENU WHERE CODE="+str(modify_item))
    a = c.fetchall()
    if a[0][0] == 0:
      print("Item not in menu.")
    else:
      modify_price = int(input("Price of item:"))
      c.execute("UPDATE Menu SET PRICE ="+str(modify_price)+" WHERE CODE ="+str(modify_item))
      print("Item modified successfully!")
      c.execute("SELECT * FROM MENU")
      items = c.fetchall()
      for i in items:
          print(str(i[0]).ljust(10, " "), str(i[1]).ljust(15, " "), str(i[2]).ljust(35, " "))
  except:
    print("Please enter the right code.")
    conn.commit()
```

```python
#SEARCH
import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()
def searchitem():
  try:
    search_item = int(input("Code of item to to search:"))
    c.execute("SELECT COUNT(CODE) FROM MENU WHERE CODE="+str(search_item))
    a = c.fetchall()
    if a[0][0] == 0:
      print("Item not in menu.")
    else:
      c.execute("SELECT * FROM Menu WHERE CODE ="+str(search_item))
      items = c.fetchall()
      for i in items:
          print("CODE".ljust(10, " "), "ITEM".ljust(15, " "), "COST".ljust(35, " "))
          print(str(i[0]).ljust(10, " "), str(i[1]).ljust(15, " "), str(i[2]).ljust(35, " "))
  except:
    print("Please enter the right code.")
    conn.commit()
```

```python
#PRINTING A BILL

import sqlite3
conn = sqlite3.connect("Menu.db")
c = conn.cursor()
def printbill():
  bill = []
  cnt = 0
  total_price = 0
  while True:
    buyitem = input('''Code of item to add to cart
Press enter when you are done: ''')
    if buyitem == "":
      break
    c.execute("SELECT COUNT(CODE) FROM MENU WHERE CODE="+buyitem)
    a = c.fetchall()
    if a[0][0] == 0:
      print("Item not in menu.")
    else:
      try:
         qtyitem = int(input("Qty of item:"))
         c.execute("SELECT Price from MENU WHERE CODE="+buyitem)
         priceperitem=c.fetchall()
         priceitem = (priceperitem[0][0]*qtyitem)
         total_price += priceitem
         c.execute("SELECT ITEM FROM MENU WHERE CODE="+buyitem)
         itemname = c.fetchall()
         buyitem_name = (itemname[0][0])
         bill.append((cnt+1,buyitem_name, qtyitem, priceitem))
         cnt += 1
      except:
        print("Something went wrong, please try again.")
  print("*********************************************************************")
  print()
  print("BILL")
  input("Name:")
  input("Date:")
  print("CODE".ljust(10, " "), "Item".ljust(15, " "), "Qty".ljust(35, " "), "Price".ljust(40," "))
  for i in bill:
    print(str(i[0]).ljust(10, " "), str(i[1]).ljust(15, " "), str(i[2]).ljust(35, " "),str(i[3]).ljust(40," "))
  print("Your total price is", total_price)
  print("Thank you for your purchase!")
  print()
  print("*********************************************************************")
  conn.commit()
```
